```python
dataframes = []

for file_path in file_paths:
  url = f"https://raw.githubusercontent.com/{username}/{repo_name}/{branch_name}/{file_path}"

  # Send a request to fetch the CSV file with authorization
  headers = {'Authorization': f'token {token}'}
  response = requests.get(url, headers=headers)

  # Check if the request was successful
  if response.status_code == 200:
      # Read the CSV file into a DataFrame
      data = StringIO(response.text)
      df = pd.read_csv(data)
      dataframes.append(df)
      print(f"Successfully fetched data from {file_path}")
  else:
      print(f"Failed to fetch data: {response.status_code}")

print()
#concate all data frame
merged_data = pd.concat(dataframes, ignore_index=True)

print("Total shape: ", merged_data.shape)
print()
print(merged_data.head())
```

```python
# Get numeric columns
numeric_columns = df.select_dtypes(include='number').columns

# Create interactive distribution plots with different colors for each column
for idx, col in enumerate(numeric_columns):
    # Select a color scale
    color_scale = px.colors.qualitative.Plotly[idx % len(px.colors.qualitative.Plotly)]  # Rotate colors

    # Create a colorful histogram with marginal box plot
    fig = px.histogram(df, x=col, nbins=30, marginal="box", title=f"Distribution of {col}",
                       color_discrete_sequence=[color_scale], template="plotly_dark")

    # Update appearance
    fig.update_traces(opacity=0.8)
    fig.update_layout(
        xaxis_title=col,
        yaxis_title="Count",
        title_font_size=18,
        title_x=0.5,
        bargap=0.2,
        paper_bgcolor="rgba(0,0,0,0)",
        plot_bgcolor="rgba(0,0,0,0)",
    )

    #fig.show()
```

```python
# List of numeric columns
numeric_columns = df.select_dtypes(include='number').columns

# Dropdown widget to select column
dropdown = widgets.Dropdown(
    options=numeric_columns,
    value=numeric_columns[0],
    description='Select Column:',
)

# Function to update the plot based on selected column
def update_plot(column):
    fig = go.Figure()

    # Add histogram
    fig.add_trace(
        go.Histogram(
            x=df[column],
            nbinsx=30,
            marker_color=px.colors.qualitative.Plotly[0],
            opacity=0.75
        )
    )

    # Update layout
    fig.update_layout(
        title=f"Distribution of {column}",
        xaxis_title=column,
        yaxis_title="Count",
        template="plotly_white",
        bargap=0.2,
    )

    fig.show()

# Display initial plot
update_plot(dropdown.value)

# Link the dropdown to the update function
dropdown.observe(lambda change: update_plot(change['new']), names='value')
display(dropdown)
```
